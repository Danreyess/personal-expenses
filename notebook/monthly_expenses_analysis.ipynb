```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("../data/gastos_mensuales.csv")
df.head()
                
```

```python
#Total monthly expenses 
expense_columns = [
    'Rent', 'Food_Groceries', 'Dining_Out', 'Transportation',
    'Utilities', 'Internet_Phone', 'Insurance', 'Gym',
    'Subscriptions', 'Entertainment', 'Travel', 'Clothing',
    'Healthcare', 'Miscellaneous'
]

df["Total_expenses"] = df[expense_columns].sum (axis = 1)
```

```python
df[["Month", "Total_expenses"]]
```

```python
#Data visualization
df.set_index("Month")["Total_expenses"].plot(kind="line", figsize=(9,5), title="Total Monthly Expenses")
plt.show()
```
